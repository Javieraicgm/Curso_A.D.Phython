```python
sentence = """La selección de Chile ganó la Copa América el año 2015 en Chile y el año 2016 en EEUU."""
token_sequence = str.split(sentence)
vocab = sorted(set(token_sequence))
print(vocab)
```

```python
#El numero de tokens en la oracion
num_tokens = len(token_sequence)
print(num_tokens)

#La cantidad de tokens en el léxico
vocab_size = len(vocab)
print(vocab_size)
```

## Vectorización de un documento

* **One Hot Vectors**


* **Bolsa de Palabras Binaria**

```python
import numpy as np

onehot_vectors = np.zeros((num_tokens, vocab_size), int)

for i, word in enumerate(token_sequence):
    onehot_vectors[i, vocab.index(word)] = 1

' '.join(vocab)

print(onehot_vectors)
print(onehot_vectors.shape)
```

```python
#Representación vectorial de una frase
rep_vec=np.transpose(onehot_vectors)
print(rep_vec)
print(rep_vec.shape)
```

```python
import pandas as pd

df = pd.DataFrame(onehot_vectors, columns=vocab)
df
```

```python
df = pd.DataFrame(onehot_vectors, columns=vocab)
df[df == 0] = ''
df
```

```python
sentences = """Chile ganó la Copa América los años 2015 y 2016.\n"""
sentences += """Chile no fué al mundial.\n"""
sentences += "En los partidos de fútbol no hay público en las galerias.\n"
sentences += """En el 2015 y 2016 no estaba en Chile.\n"""
sentences += """Puede que Chile vaya al mundial"""
corpus = {}

for i, sent in enumerate(sentences.split('\n')):
    corpus['sent{}'.format(i)] = dict((tok, 1) for tok in sent.split())
df = pd.DataFrame.from_records(corpus).fillna(0).astype(int).T
df
```
